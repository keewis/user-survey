```python
prefix = "Which of Xarray's core features are most important to you?"
df_q = df.filter(regex=f'^{prefix}',axis=1)

for k, s in df_q.items():
    plt.figure()
    s.value_counts().sort_index().plot.bar()
    plt.title(k.replace('?', '?\n'))
```

```python
from collections import defaultdict
```

```python
d = defaultdict(int)
for s in df['What are some other libraries that you often use with Xarray? (e.g. "dask, pandas, scikit-learn") '].dropna().values:
        
    for pkg in s.split(','):
        pkg = pkg.lower().strip()
        if pkg:
            d[pkg] += 1

pkgs = pd.Series(d)
pkgs = pkgs[pkgs > 5].sort_values()
pkgs.plot.bar()
```
